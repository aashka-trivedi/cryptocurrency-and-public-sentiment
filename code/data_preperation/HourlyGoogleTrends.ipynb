# Hourly Google Trend Data

## Convert Daily Data to Duplicated Hourly Data

```python
import pandas as pd
```

```python
google = pd.read_csv('../../data/daily_data/google_trends_daily_keywords.csv', sep='\t')
google.head()
```

```python
google = google.loc[:, ~google.columns.str.contains('^Unnamed')]
google.head()
```

```python
data=[]
for index,row in google.iterrows():
    rowCopy = row.copy()
    for i in range(10):
        new = rowCopy.copy()
        new['date'] = new['date'] + ' 0'+str(i)+':00:00'
        data.append(new)
    for i in range(10,25):
        new = rowCopy.copy()
        new['date'] = new['date'] + ' '+str(i)+':00:00'
        data.append(new)
```

```python
google_hours = pd.DataFrame(data)
```

```python
google_hours
```

```python
#Save as CSV
google_hours.to_csv('../../data/hourly_data/google_trends_hourly.csv', sep='\t', index=False)
```

## Get Average Hourly Google Trend Score

```python
import pandas as pd
```

```python
google = pd.read_csv('../../data/hourly_data/google_trends_hourly.csv', sep='\t', index_col= None)
```

```python
google.head()
```

```python
cols = google.columns
google['avg_score'] = google[cols[1:]].mean(axis=1)
google.head()
```

```python
google.to_csv('../../data/hourly_data/trends_hourly_score.csv', sep='\t', index=False)
```
