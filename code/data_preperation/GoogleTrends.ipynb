# Google Trends

Here, we find the scaled google search numbers for a constellation of terms relating to "bitcoin" and "btc"

References:

- https://github.com/GeneralMills/pytrends/blob/master/pytrends/dailydata.py

```python
import pandas as pd                        
from pytrends.request import TrendReq
from pytrends import dailydata
```

```python
#Connect to Google 
pytrends = TrendReq()

#Get Interest over time 
keywords = ['bitcoin']
#IMPORTANT: Use specific timeframe- using "now" or "today" throws a 400 error
pytrends.build_payload(kw_list=keywords, cat=0, timeframe='2021-01-01 2021-08-31', geo='', gprop='')
topics = pytrends.related_queries()
```

```python
topics
```

```python
#Get Interest over time 
keywords = ['btc']
#IMPORTANT: Use specific timeframe- using "now" or "today" throws a 400 error
pytrends.build_payload(kw_list=keywords, cat=0, timeframe='2021-01-01 2021-08-31', geo='', gprop='')
topics = pytrends.related_queries()

topics
```

```python
keywords = ['bitcoin', 'btc', 'bitcoin price', 'bitcoin kurs', 'bitcoin usd', 'bitcoin stock', 'bitcoin dollar', 'bitcoin euro', 'buy bitcoin', 'buy btc']
```

```python
daily_data = pd.DataFrame()
```

```python
for i,topic in enumerate(keywords):
    df = dailydata.get_daily_data(topic, 2021, 1, 2021, 8, geo = '')
    df = df.reset_index()
    if i==0:
        daily_data['date'] = df['date'] 
    daily_data[topic] = df[topic]
```

```python
daily_data
```

```python
keywords_btc = ['btc usd', 'btc inr', 'price btc', 'btc stock', 'btc coin', 'btc euro']
```

```python
for topic in keywords_btc:
    df = dailydata.get_daily_data(topic, 2021, 1, 2021, 8, geo = '')
    df = df.reset_index()
    daily_data[topic] = df[topic]
```

```python
daily_data.head()
```

```python
data_path = 'data/google_trends_daily_keywords.csv'
daily_data.to_csv(data_path, sep='\t')
```

## Attempt 2- Using weighted average 

Weight of each score is proportional to how much that term is searched for on average

```python
data = pd.read_csv('data/google_trends_daily_keywords.csv', sep='\t', index_col=0)
data.head()
```

```python
#Obtain Weights
col_sum = data.mean(axis=0, skipna = True, numeric_only=True)
cnormalized_sum = col_sum/col_sum.sum(axis=0)
normalized_sum
```

```python
#Find Weighted Sum
cols = data.columns[1:]
scores = []
for index, row in data.iterrows():
    score = 0
    for col in cols:
        score = score + normalized_sum[col]*row[col]
    scores.append(score)
data['weighted_score'] = scores
data.head()
```

```python
#Make csv file

new_data['weighted_score'] = data['weighted_score']
new_data.head()
```

```python
new_data.to_csv('data/trends_daily_score.csv', sep='\t')
```
