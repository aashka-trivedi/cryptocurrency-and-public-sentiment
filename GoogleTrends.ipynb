```python
import pandas as pd                        
from pytrends.request import TrendReq
from pytrends import dailydata
```

```python
#Connect to Google 
pytrends = TrendReq()

#Get Interest over time 
keywords = ['bitcoin']
#IMPORTANT: Use specific timeframe- using "now" or "today" throws a 400 error
pytrends.build_payload(kw_list=keywords, cat=0, timeframe='2021-01-01 2021-08-31', geo='', gprop='')
topics = pytrends.related_queries()
```

```python
keywords = ['bitcoin', 'btc', 'bitcoin price', 'bitcoin kurs', 'bitcoin usd', 'bitcoin stock', 'bitcoin dollar', 'bitcoin euro', 'buy bitcoin', 'buy btc']
```

```python
daily_data = pd.DataFrame()
```

```python
for i,topic in enumerate(keywords):
    df = dailydata.get_daily_data(topic, 2021, 1, 2021, 8, geo = '')
    df = df.reset_index()
    if i==0:
        daily_data['date'] = df['date'] 
    daily_data[topic] = df[topic]
```

```python
daily_data
```

```python
keywords_btc = ['btc usd', 'btc inr', 'price btc', 'btc stock', 'btc coin', 'btc euro']
```

```python
for topic in keywords_btc:
    df = dailydata.get_daily_data(topic, 2021, 1, 2021, 8, geo = '')
    df = df.reset_index()
    daily_data[topic] = df[topic]
```
